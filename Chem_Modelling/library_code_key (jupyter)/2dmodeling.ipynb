```python
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem import Descriptors
from rdkit.Chem import AllChem
from rdkit import DataStructs
import numpy as np
```

How to generate a chem molecule from SMILES code (generates image on command-no need to import PIL or additional image processing systems)

```python
mol = Chem.MolFromSmiles('CCCC')
mol
```

Convert back to SMILES (learn to incorporate a database scraping for this (pubchem))

```python
smiles = Chem.MolToSmiles(mol)
smiles
```

You can also access molecular attributes using the Descriptor library

```python
mw = Descriptors.MolWt(mol)
mw
```

```python
# List of molecules from smiles 
# (generating several images)

smiles_list = ['CC1CCC/C(C)=C1/C=C/C(C)=C/C=C/C(C)=C/C=C/C=C(C)/C=C/C=C(C)/C=C/C2=C(C)/CCCC2(C)C', 'CCO', 'F/C=C\F']

mol_list = []
for smiles in smiles_list:
    mol = Chem.MolFromSmiles(smiles)
    mol_list.append(mol)

# Draws a series of mols from a list (additional parameters specifies how many per row too)
img = Draw.MolsToGridImage(mol_list, molsPerRow=1)
img
```

Checks if a specific molecule (in SMILES format) is present in a list of different molecules

```python
pattern = Chem.MolFromSmiles('S')
for mol in mol_list:
    print(mol.HasSubstructMatch(pattern))
```
